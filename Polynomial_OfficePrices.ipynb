```python
import numpy as np
from sklearn import linear_model as lm
from sklearn.preprocessing import PolynomialFeatures as pp

# Read F and H from input
F, H = map(int, input().split())

# Read the training data
X_train = np.array([input().split() for _ in range(H)], dtype=float)
train = X_train  # Renaming for clarity

# Read T and the test data
T = int(input())
X_test = np.array([input().split() for _ in range(T)], dtype=float)

# Separate the input features and the target variable
X_train_features = train[:, :-1]
X_train_target = train[:, -1]

# Create a PolynomialFeatures transformer
poly = pp(degree=3, include_bias=False)

# Transform the training data
X_train_poly = poly.fit_transform(X_train_features)

# Initialize the linear regression model
model = lm.LinearRegression()

# Fit the model
model.fit(X_train_poly, X_train_target)

# Transform the test data using the same PolynomialFeatures
X_test_poly = poly.transform(X_test)

# Make predictions
Y = model.predict(X_test_poly)

# Print the predictions
for val in Y:
    print("{:.2f}".format(val))
```
